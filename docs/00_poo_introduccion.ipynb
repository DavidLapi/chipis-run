# ğŸ—ï¸ Tu Primera Clase: Jugador

Â¡Hora de crear tu primera clase! Una **clase** es como un **molde para galletas** ğŸª - defines la forma una vez, y puedes hacer todas las galletas que quieras.

## ğŸ­ De problema real a soluciÃ³n POO

**Problema**: Necesitamos muchos personajes en nuestro juego  
**SoluciÃ³n**: Â¡Crear una clase `Jugador`!

````python
# âœ¨ Â¡Nuestra primera clase!
class Jugador:
    def __init__(self, nombre):
        """
        Constructor: se ejecuta cuando creamos un nuevo jugador
        """
        print(f"ğŸ® Â¡Creando jugador {nombre}!")
        
        # ğŸ“¦ ATRIBUTOS: caracterÃ­sticas del jugador
        self.nombre = nombre
        self.vidas = 3
        self.puntos = 0
        self.x = 100  # posiciÃ³n horizontal
        self.y = 200  # posiciÃ³n vertical
        
        print(f"âœ… {self.nombre} estÃ¡ listo con {self.vidas} vidas")
    
    # âš¡ MÃ‰TODOS: acciones que puede hacer el jugador
    def mover(self, direccion):
        """Mover el jugador en una direcciÃ³n"""
        if direccion == "izquierda":
            self.x -= 10
            print(f"ğŸ‘ˆ {self.nombre} se moviÃ³ a la izquierda (x={self.x})")
        elif direccion == "derecha":
            self.x += 10
            print(f"ğŸ‘‰ {self.nombre} se moviÃ³ a la derecha (x={self.x})")
        elif direccion == "arriba":
            self.y -= 10
            print(f"ğŸ‘† {self.nombre} subiÃ³ (y={self.y})")
        elif direccion == "abajo":
            self.y += 10
            print(f"ğŸ‘‡ {self.nombre} bajÃ³ (y={self.y})")
    
    def puntuar(self, puntos):
        """AÃ±adir puntos al jugador"""
        self.puntos += puntos
        print(f"ğŸŒŸ Â¡{self.nombre} ganÃ³ {puntos} puntos! Total: {self.puntos}")
    
    def herir(self):
        """Quitar una vida al jugador"""
        if self.vidas > 0:
            self.vidas -= 1
            print(f"ğŸ’” Â¡{self.nombre} perdiÃ³ una vida! Quedan: {self.vidas}")
            if self.vidas == 0:
                print(f"ğŸ’€ Â¡{self.nombre} ha perdido el juego!")
        return self.vidas > 0  # True si sigue vivo

print("ğŸ¯ Â¡Clase Jugador creada! Ahora vamos a usarla...")
````

## ğŸª Â¡Hora del espectÃ¡culo! Creemos personajes

Ejecuta la celda de arriba y luego esta:

````python
# ğŸ­ Creando OBJETOS desde la clase (como galletas del molde)
print("=== ğŸ­ FÃBRICA DE JUGADORES ===")

# Crear dos jugadores diferentes
julia = Jugador("Julia")
cale = Jugador("Cale")

print("\n=== ğŸ® ESTADO INICIAL ===")
print(f"ğŸ“Š {julia.nombre}: {julia.vidas} vidas, {julia.puntos} puntos, posiciÃ³n ({julia.x}, {julia.y})")
print(f"ğŸ“Š {cale.nombre}: {cale.vidas} vidas, {cale.puntos} puntos, posiciÃ³n ({cale.x}, {cale.y})")

print("\nğŸ’¡ Â¡FÃ­jate! Son dos OBJETOS diferentes con el mismo 'molde' (clase)")
print(f"ğŸ” julia es de tipo: {type(julia)}")
print(f"ğŸ” cale es de tipo: {type(cale)}")
````

**Roxi**: "Â¡Oye, muÃ©vete a la derecha!"

**En POO es igual**: Cuando llamas `mover("derecha")`, Â¿quiÃ©n se mueve?

````python
# ğŸ” Â¡Experimento para entender self!
print("=== ğŸ§ª EXPERIMENTO SELF ===")

# Creamos varios estudiantes
miguel = Jugador("Miguel")
marcus = Jugador("Marcus")
dani = Jugador("Dani")

print("\n=== ğŸ¯ POSICIONES INICIALES ===")
print(f"Miguel estÃ¡ en x={miguel.x}")
print(f"Marcus estÃ¡ en x={marcus.x}")  
print(f"Dani estÃ¡ en x={dani.x}")

print("\n=== ğŸƒ Â¡CADA UNO SE MUEVE INDEPENDIENTEMENTE! ===")

# Cada uno se mueve - Â¡observa los mensajes!
miguel.mover("derecha")  # self serÃ¡ miguel
marcus.mover("izquierda")   # self serÃ¡ marcus  
dani.mover("derecha")    # self serÃ¡ dani

print("\n=== ğŸ“ POSICIONES FINALES ===")
print(f"Miguel terminÃ³ en x={miguel.x}")
print(f"Marcus terminÃ³ en x={marcus.x}")
print(f"Dani terminÃ³ en x={dani.x}")

print("\nğŸ’¡ Â¡FÃ­jate! Cada objeto modificÃ³ SU PROPIO estado, no el de los otros")
print("ğŸ¯ ESO es lo que hace 'self': indica 'mi propio estado'")
````

## ğŸ® Mini-Ejercicio: Â¡AÃ±ade velocidad!

**Tu misiÃ³n**: Mejora la clase `Jugador` aÃ±adiendo:
- Un atributo `velocidad` (por defecto 10)  
- Un mÃ©todo `acelerar(factor)` que multiplique la velocidad
- Modifica `mover()` para usar la velocidad actual

**Â¡Piensa!** Â¿DÃ³nde va cada cosa?

````python
# ğŸš€ Â¡Mejoremos la clase Jugador!
class JugadorMejorado:
    def __init__(self, nombre):
        print(f"ğŸ® Â¡Creando jugador mejorado {nombre}!")
        
        # Atributos existentes
        self.nombre = nombre
        self.vidas = 3
        self.puntos = 0
        self.x = 100
        self.y = 200
        
        # ğŸ†• Â¡NUEVO! Atributo velocidad
        self.velocidad = 10  # pÃ­xeles por movimiento
        
        print(f"âœ… {self.nombre} estÃ¡ listo con velocidad {self.velocidad}")
    
    def mover(self, direccion):
        """Mover usando la velocidad actual"""
        if direccion == "izquierda":
            self.x -= self.velocidad  # Â¡Usa self.velocidad!
            print(f"ğŸ‘ˆ {self.nombre} se moviÃ³ {self.velocidad} pÃ­xeles a la izquierda (x={self.x})")
        elif direccion == "derecha":
            self.x += self.velocidad
            print(f"ğŸ‘‰ {self.nombre} se moviÃ³ {self.velocidad} pÃ­xeles a la derecha (x={self.x})")
        # ... (arriba y abajo igual)
    
    def acelerar(self, factor):
        """ğŸš€ Acelerar el jugador"""
        self.velocidad *= factor
        print(f"ğŸƒâ€â™€ï¸ Â¡{self.nombre} ahora va a velocidad {self.velocidad}!")
    
    def puntuar(self, puntos):
        self.puntos += puntos
        print(f"ğŸŒŸ Â¡{self.nombre} ganÃ³ {puntos} puntos! Total: {self.puntos}")
    
    def herir(self):
        if self.vidas > 0:
            self.vidas -= 1
            print(f"ğŸ’” Â¡{self.nombre} perdiÃ³ una vida! Quedan: {self.vidas}")
        return self.vidas > 0

print("ğŸ¯ Â¡Clase mejorada creada!")
````

````python
# ğŸ Â¡Carrera de velocidad entre compaÃ±eros!
print("=== ğŸ CARRERA DE VELOCIDAD ===")

# Creamos corredores de la clase
presi = JugadorMejorado("Presi")
ramazan = JugadorMejorado("Ramazan") 
wizi = JugadorMejorado("Wizi")

print("\n=== ğŸš€ Â¡MODIFICAMOS VELOCIDADES! ===")
presi.acelerar(2)        # Presi va 2x mÃ¡s rÃ¡pido
ramazan.acelerar(1.5)    # Ramazan 1.5x mÃ¡s rÃ¡pido  
# Wizi mantiene velocidad normal

print("\n=== ğŸƒ Â¡A CORRER! ===")
print("ğŸ Cada uno da 3 pasos a la derecha...")

for paso in range(1, 4):
    print(f"\n--- Paso {paso} ---")
    presi.mover("derecha")
    ramazan.mover("derecha")
    wizi.mover("derecha")

print("\n=== ğŸ† RESULTADOS FINALES ===")
print(f"ğŸ¥‡ Presi: posiciÃ³n x={presi.x} (velocidad {presi.velocidad})")
print(f"ğŸ¥ˆ Ramazan: posiciÃ³n x={ramazan.x} (velocidad {ramazan.velocidad})")
print(f"ğŸ¥‰ Wizi: posiciÃ³n x={wizi.x} (velocidad {wizi.velocidad})")

print("\nğŸ’¡ Â¡Cada objeto mantiene SU PROPIA velocidad!")
print("ğŸ¯ Esto es encapsulaciÃ³n: cada objeto controla su propio estado")
````

# ğŸ”« ComposiciÃ³n Simple: Â¡Objetos que usan otros objetos!

**Nuevo reto**: Nuestros jugadores necesitan **cuchillos** para defenderse.

Â¿CÃ³mo modelamos esto? ğŸ¤”

## ğŸ§© SoluciÃ³n: Â¡Una clase usa otra clase!

- **Jugador** â†’ puede crear y usar **Cuchillos**
- **Cuchillo** â†’ tiene su propia lÃ³gica (direcciÃ³n, velocidad)

**Esto es COMPOSICIÃ“N**: Un objeto "tiene" otros objetos

````python
# ğŸ”ª Â¡Creemos la clase Cuchillo!
class Cuchillo:
    def __init__(self, x, y, direccion):
        """Un cuchillo tiene posiciÃ³n y direcciÃ³n"""
        self.x = x
        self.y = y
        self.direccion = direccion
        self.velocidad = 15  # Los cuchillos van rÃ¡pido
        self.activo = True   # Â¿Sigue volando?
        
        print(f"ğŸ”ª Â¡Cuchillo lanzado desde ({x}, {y}) hacia {direccion}!")
    
    def update(self):
        """Mover el cuchillo segÃºn su direcciÃ³n"""
        if self.activo:
            if self.direccion == "arriba":
                self.y -= self.velocidad
            elif self.direccion == "abajo":  
                self.y += self.velocidad
            elif self.direccion == "izquierda":
                self.x -= self.velocidad
            elif self.direccion == "derecha":
                self.x += self.velocidad
            
            # Â¿Se saliÃ³ de la pantalla?
            if self.y < 0 or self.y > 400 or self.x < 0 or self.x > 800:
                self.activo = False
                print(f"ğŸ’¨ Cuchillo saliÃ³ de pantalla en ({self.x}, {self.y})")

# ğŸ® Â¡Jugador que puede lanzar cuchillos!
class JugadorConCuchillos:
    def __init__(self, nombre):
        self.nombre = nombre
        self.x = 100
        self.y = 200
        self.cuchillos = []  # ğŸ“¦ Lista de cuchillos lanzados
        self.enfriamiento = 0  # Contador para evitar spam
        
        print(f"ğŸ® {self.nombre} estÃ¡ listo para lanzar cuchillos!")
    
    def lanzar_cuchillo(self, direccion):
        """ğŸ”« Lanzar un cuchillo si no hay enfriamiento"""
        if self.enfriamiento <= 0:
            # Crear un nuevo cuchillo en mi posiciÃ³n
            nuevo_cuchillo = Cuchillo(self.x, self.y, direccion)
            self.cuchillos.append(nuevo_cuchillo)
            self.enfriamiento = 3  # 3 turnos de enfriamiento
            print(f"ğŸ¯ {self.nombre} lanzÃ³ un cuchillo hacia {direccion}")
        else:
            print(f"â° {self.nombre} debe esperar {self.enfriamiento} turnos")
    
    def update(self):
        """Actualizar enfriamiento y todos los cuchillos"""
        # Reducir enfriamiento
        if self.enfriamiento > 0:
            self.enfriamiento -= 1
        
        # Actualizar todos los cuchillos
        for cuchillo in self.cuchillos:
            cuchillo.update()
        
        # Limpiar cuchillos inactivos
        cuchillos_activos = [c for c in self.cuchillos if c.activo]
        eliminados = len(self.cuchillos) - len(cuchillos_activos)
        self.cuchillos = cuchillos_activos
        
        if eliminados > 0:
            print(f"ğŸ§¹ {self.nombre} limpiÃ³ {eliminados} cuchillos inactivos")

print("ğŸ¯ Â¡Clases de combate creadas!")
````

````python
# âš”ï¸ Â¡Duelo de cuchillos entre RubÃ©n y David!
print("=== âš”ï¸ DUELO DE CUCHILLOS ===")

ruben = JugadorConCuchillos("RubÃ©n")
david = JugadorConCuchillos("David")

print("\n=== ğŸ¯ RONDA 1: LANZAMIENTOS ===")
ruben.lanzar_cuchillo("arriba")
david.lanzar_cuchillo("derecha")
ruben.lanzar_cuchillo("izquierda")  # Â¿PodrÃ¡ lanzar tan rÃ¡pido?

print(f"\nğŸ“Š RubÃ©n tiene {len(ruben.cuchillos)} cuchillos volando")
print(f"ğŸ“Š David tiene {len(david.cuchillos)} cuchillos volando")

print("\n=== â° SIMULANDO 5 TURNOS ===")
for turno in range(1, 6):
    print(f"\n--- Turno {turno} ---")
    
    # Actualizar ambos jugadores
    ruben.update()
    david.update()
    
    # Intentar lanzar mÃ¡s cuchillos
    if turno == 2:
        david.lanzar_cuchillo("arriba")
    if turno == 4:
        ruben.lanzar_cuchillo("derecha")
        david.lanzar_cuchillo("izquierda")
    
    print(f"ğŸ“Š Estado turno {turno}:")
    print(f"   RubÃ©n: {len(ruben.cuchillos)} cuchillos, enfriamiento: {ruben.enfriamiento}")
    print(f"   David: {len(david.cuchillos)} cuchillos, enfriamiento: {david.enfriamiento}")

print("\nğŸ’¡ Â¡FÃ­jate los conceptos de POO!")
print("ğŸ”¸ COMPOSICIÃ“N: Jugador 'tiene' una lista de Cuchillos")
print("ğŸ”¸ RESPONSABILIDADES: Cada clase sabe quÃ© hacer")
print("ğŸ”¸ ESTADO: Cada objeto mantiene sus propios datos")
````
